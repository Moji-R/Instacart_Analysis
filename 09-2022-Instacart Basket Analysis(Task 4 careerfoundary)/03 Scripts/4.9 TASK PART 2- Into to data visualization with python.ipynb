#### 2b: creating sampling data and line chart

```python
# create a list holding True/False values to the test np.random.rand() <= 0.7
np.random.seed(4)
dev = np.random.rand(len(ords_prods_cus)) <=0.7
```

```python
# Printing dev
dev
```

```python
# store 70% of the sample in dataframe big
big = ords_prods_cus[dev]
```

```python
#store 30% of the sample in dataframe small
small = ords_prods_cus[~dev]
```

```python
# Checking length of ords_prods_cus dataframe
len(ords_prods_cus)
```

```python
# checking length of "big" sample data
len(big)
```

```python
# checking length of small sample
len(small)
```

```python
# checking length of all sample
len(big) + len(small)
```
