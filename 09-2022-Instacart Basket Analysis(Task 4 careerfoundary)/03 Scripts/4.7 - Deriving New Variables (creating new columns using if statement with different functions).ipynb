# 04. If and user-define function

```python
# Define a function
def price_label(row):
    if row['prices']<=5:
        return 'Low range product'
    elif (row['prices']>5) and(row['prices']<= 15):
        return 'Mid range products'
    elif row['prices']>15:
        return 'High range products'
    else: return np.nan
```

```python
# Apply the function
df['price_range'] = df.apply(price_label,axis=1)
```

```python
# counting values
df['price_range'].value_counts(dropna = False)
```

```python
# Checking the maximum value
df['prices'].max()
```

# 05. Using the if and loc function

```python
# using loc function( prices > 15 is categorized as High range product)
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Prices <= 15 & prices > 5 is categorized as Mid range product
df.loc[(df['prices'] <=15) & (df['prices']>5), 'price_range_loc'] = 'Mid-range product'
```

```python
# prices <=5 is categorized as Low range product
df.loc[df['prices'] <= 5,'price_range_loc'] = 'Low-range product'
```

```python
# counting values
df['price_range_loc'].value_counts(dropna= False)
```
